```python
def get_high_mask(arr):
    # Pad outside 
    data_p = np.pad(arr, 1, "constant", constant_values=9)

    # Check adjacent
    u = (data_p[1:] >= data_p[:-1])[:-1, 1:-1]
    d = (data_p[:-1] >= data_p[1:])[1:, 1:-1]
    l = (data_p[:, 1:] >= data_p[:, :-1])[1:-1, :-1]
    r = (data_p[:, :-1] >= data_p[:, 1:])[1:-1, 1:]

    return u & d & l & r

def get_size(arr, start):
    """Gets the size of a basin, all connected to "start" that are True """
    arr_p = np.pad(arr, 1, "constant", constant_values=False)

    size = 0
    stack = [start]
    visited = set()
    while len(stack) > 0:
        i, j = stack.pop()
        visited.add((i, j))

        pad_i, pad_j = i + 1, j + 1
        
        if arr_p[pad_i, pad_j]:
            size += 1
            arr[i, j] = False
            # Add any adjacent onto the stack
            new_elements = [(i+1, j), (i-1, j), (i, j+1), (i, j-1)]
            stack.extend(new_elements)
            stack = [x for x in stack if x not in visited]

    return arr, size

def get_basins(data):
    data = np.asarray([[y for y in x] for x in data.split("\n")], dtype="int")

    # data = ~get_high_mask(data)
    data = data != 9

    sizes = []
    for i, j in np.ndindex(data.shape):
        data, size = get_size(data, (i, j))
        if size > 0:
            sizes.append(size)

    sizes = list(sorted(sizes))
    print(sizes)
    print(np.product(sizes[-3:]))

get_basins(test_data)
```

```python
SCORES = {
    ")": 1,
    "]": 2,
    "}": 3,
    ">": 4
}

def complete_syntax(data):
    data = data.split("\n")
    scores = []
    for line in data:
        corrupt = False
        score = 0
        stack = []
        for x in line:
            if x in OPENS:
                stack.append(x)
            else:
                y = stack.pop()
                if OPENS.index(y) != CLOSES.index(x):
                    corrupt = True
                    break
        if corrupt:
            continue
        # Only incomplete lines now
        for x in [SCORES[CLOSES[OPENS.index(x)]] for x in reversed(stack)]:
            score = score * 5 + x
        scores.append(score)
    scores = list(sorted(scores))
    print(scores[(len(scores) - 1) // 2])
complete_syntax(test_data)
```

```python
from itertools import product


def octo_step(arr):

    flashes = 0
    arr += 1
    flashed = []

    arr = np.pad(arr, 1, "constant", constant_values=-3)

    stack = np.argwhere(arr > 9).tolist()
    while len(stack) > 0:
        i, j = stack.pop()
        try:
            arr[i, j] = 0
            flashes += 1
        except IndexError:
            continue
        # flash adjacent
        adjacent = np.asarray([
            [a, b] for a, b in product(range(i-1, i+2), range(j-1, j+2))
            if not (a == i and b == j)
        ])
        arr[adjacent[:, 0], adjacent[:, 1]] += ((arr[adjacent[:, 0], adjacent[:, 1]] != 0) * 1)

        if len(stack) == 0:
            stack = np.argwhere(arr > 9).tolist()

    arr = arr[1:-1, 1:-1]
    
    return arr, flashes

def print_board(arr):
    for x in arr:
        print(x)
    print()

def count_steps(data, steps):

    data = np.asarray([[y for y in x ]for x in data.split("\n")], dtype=int)

    print("Initial Board")
    print_board(data)

    flashes = 0
    for step in range(steps):
        data, num_flash = octo_step(data)
        flashes += num_flash

    print("Final Board")
    print_board(data)

    print(flashes, "\n")

count_steps(test_data, 100)
```

```python
from collections import defaultdict
from dataclasses import dataclass
from typing import Dict, List


@dataclass
class Node:
    id: str
    visited: bool = False
    adjacent: List[str] = None

    def __init__(self, id, visited, adjacent) -> None:
        self.id = id
        self.visited = visited
        self.adjacent = adjacent or None

    def __str__(self) -> str:
        return f'{"-" if self.visited else ""}{self.id} - {self.adjacent}'

class Nodes:

    def __init__(self, nodes=None):
        self.nodes = nodes or {}
    def get_node(self, node_id) -> Node:
        return self.nodes.get(node_id)

    def add_node(self, node: Node):
        n = self.get_node(node.id)
        if n is None:
            self.nodes[node.id] = node
        else:
            self.nodes[node.id].adjacent.append(node.adjacent[0])

    def __str__(self) -> str:
        s = ""
        for x in self.nodes.values():
            s += str(x) + "\n"
        return s
            

def build_graph(data):
    data = [x.split("-") for x in data.split("\n")]
    nodes = Nodes()
    for line in data:
        a, b = line
        nodes.add_node(Node(a, False, [b]))
        nodes.add_node(Node(b, False, [a]))

    print(nodes)
    return nodes

def build_paths(nodes: Nodes):
    # Depth first search
    node = nodes.get_node("start")
    node.visited = True
    paths = []

    stack = [["start", x] for x in node.adjacent]
    while len(stack) > 0:
        current_path = stack.pop()
        last_node = nodes.get_node(current_path[-1])
        if last_node.id == "end":
            paths.append(current_path)
        else:
            for next_node in last_node.adjacent:
                if next_node.upper() != next_node and next_node in current_path:
                    continue
                next_path = current_path + [next_node]
                stack.append(next_path)

    print(f"Found {len(paths)} paths")
    # for path in paths:
    #     print(path)
        

t1 = build_graph(test_data1)
build_paths(t1)
```

```python
import numpy as np
import pandas as pd


def evolve_polymer(data, n_steps) -> list:
    p, pairs = data.split("\n\n")
    p = [x for x in p]
    # print(len(p))
    pairs = [x.split(" -> ") for x in pairs.split("\n")]
    for i in range(n_steps):
        p_str = "".join(p)
        # Build the list of insertion
        inserts = []
        for pair, val in pairs:
            if pair not in p_str:
                continue
            
            idx = 0
            while idx < len(p_str):
                if pair not in p_str[idx:]:
                    break
                sub_idx = p_str[idx:].index(pair)
                # print(idx, pair, p_str[idx:], sub_idx)
                inserts.append((idx + sub_idx + 1, val))
                idx += sub_idx + 1
        # print(inserts)
        # loop through in reverse and insert into p
        for idx, val in reversed(sorted(inserts, key=lambda x: x[0])):
            p.insert(idx, val.lower())
        # print(i + 1, "".join(p[:30]))
        p = [x.upper() for x in p]
    return p


def calc_polymer(data, n_steps=10):
    p = evolve_polymer(data, n_steps)

    counts = [p.count(x) for x in set(p)]
    print(max(counts) - min(counts))


calc_polymer(test_data, 10)
calc_polymer(input_data, 10)
```

```python
def track_polymer(data, n_steps, verbose=False):
    p, rules = data.split("\n\n")
    start = p[0]
    polymer = defaultdict(lambda: 0)
    for i in range(len(p) - 1):
        pair = p[i] + p[i+1]
        polymer[pair] += 1

    rules = [x.split(" -> ") for x in rules.split("\n")]
    def step_polymer(polymer) -> Dict[str, int]:
        # Step the polymer
        new_polymer = defaultdict(lambda: 0)
        for pair, new_part in rules:
            occurences = polymer[pair]
            if occurences == 0:
                continue
            
            pair_a = pair[0] + new_part
            pair_b = new_part + pair[1]
            
            new_polymer[pair_a] += polymer[pair]
            new_polymer[pair_b] += polymer[pair]
        return new_polymer

    for i in range(n_steps):
        polymer = step_polymer(polymer)
        parts = set([x[1] for x in polymer])
        counts = defaultdict(lambda: 0)
        counts[start] += 1
        for pair in polymer:
            counts[pair[1]] += polymer[pair]

        counts = [x for x in counts.values()]
        score = max(counts) - min(counts)
        if verbose:
            print(counts)
            print(dict(polymer))
            print(f"length = {sum(counts)}, score = {score}")

    return score

score = track_polymer(test_data, 10)
print(score)
score = track_polymer(test_data, 40)
print(score)
score = track_polymer(input_data, 40)
print(score)
```

```python
def adjacent_idxs(idx):
    yield idx[0] + 1, idx[1]
    yield idx[0] - 1, idx[1]
    yield idx[0], idx[1] + 1
    yield idx[0], idx[1] - 1

def get_grid(data):
    grid = np.asarray([[x for x in y] for y in data.split("\n")], dtype=int)
    costs = {}
    for i in range(grid.shape[0]):
        for j in range(grid.shape[1]):
            idx = (i, j)
            num = np.ravel_multi_index(idx, grid.shape)
            costs[num] = {} 

            for adj in adjacent_idxs(idx):
                try:
                    adj_num = np.ravel_multi_index(adj, grid.shape)
                    costs[num][adj_num] = grid[adj]
                except Exception:
                    continue
    print(f"Got grid size {grid.shape}")
    return grid, costs

def shortest_path(costs):
    # print(costs)
    unvisited = [x for x in set([x for x in costs])]
    shortest = {x: np.inf for x in unvisited}
    node = min(unvisited)
    target = max(unvisited)

    start_cost = costs[2][1]
    shortest[node] = start_cost

    while target in unvisited:
        node = unvisited.pop(0)
        for adj in costs[node]:
            if adj in unvisited:
                shortest[adj] = min(shortest[adj], shortest[node] + costs[node][adj])

        unvisited.sort(key=lambda x: shortest[x])

    return shortest[target] - start_cost

grid, costs = get_grid(test_data)
target = shortest_path(costs)
print(target)
grid, costs = get_grid(input_data)
target = shortest_path(costs)
print(target)
```

```python
import heapq

def get_expanded_grid(data):
    grid = np.asarray([[x for x in y] for y in data.split("\n")], dtype=int)
    grid = np.vstack([np.hstack([grid + i for i in range(5)]) + j for j in range(5)])
    grid = np.where(9 - grid < 0, np.abs(9 - grid), grid)
    costs = {}
    for i in range(grid.shape[0]):
        for j in range(grid.shape[1]):
            idx = (i, j)
            num = np.ravel_multi_index(idx, grid.shape)
            costs[num] = {} 

            for adj in adjacent_idxs(idx):
                try:
                    adj_num = np.ravel_multi_index(adj, grid.shape)
                    costs[num][adj_num] = grid[adj]
                except Exception:
                    continue
    print(f"Got grid size {grid.shape}")
    return grid, costs

def shortest_path_heapq(costs):
    # print(costs)
    shortest = {x: np.inf for x in set([x for x in costs])}
    start_cost = costs[2][1]
    shortest[0] = start_cost
    visited = set()
    unvisited = [(start_cost, 0)]
    heapq.heapify(unvisited)
    target = max(costs)

    while len(unvisited) > 0:
        cost, node = heapq.heappop(unvisited)
        if node in visited:
            continue
        visited.add(node)
        for adj in costs[node]:
            shortest[adj] = min(shortest[adj], cost + costs[node][adj])
            if adj not in visited:
                heapq.heappush(unvisited, (shortest[adj], adj))

    return shortest[target] - start_cost
grid, costs = get_expanded_grid(test_data)
target = shortest_path_heapq(costs)
print(target)
grid, costs = get_expanded_grid(input_data)
target = shortest_path_heapq(costs)
print(target)
```
